```python
from chkp_harmony_endpoint_management_sdk import HarmonyEndpoint, InfinityPortalAuth
import os
import pandas as pd
```

```python
def load_configuration():
    client_id = os.getenv('CLIENT_ID')
    api_key = os.getenv('API_SECRET')
    api_url = os.getenv('API_URL')
    return client_id, api_key, api_url

he = HarmonyEndpoint()
client_id, api_key, api_url = load_configuration()

he.connect(infinity_portal_auth=InfinityPortalAuth(
        client_id=client_id, # O "Client ID"
        access_key=api_key, # A "Secret Key"
        gateway=api_url # A "Authentication URL"
        )) 
```

```python
print("Requisitando metadados de todas as políticas")
rules_metadata_res = he.policy_general_api.get_all_rules_metadata(header_params={ "x-mgmt-run-as-job": 'on'})
rules_metadata = rules_metadata_res.payload

rules = []
for item in rules_metadata:
    record = {
        "Orientation": item["orientation"],
        "ID": item["id"],
        "Name": item["name"],
        "Connection State": item["connectionState"],
        "Family": item["family"],
        "Is Default Rule": item["isDefaultRule"],
        "Order": item["order"]
    }
    rules.append(record)

df_rules = pd.DataFrame(rules)
df_rules
```

```python
print("Obtendo metadata apenas das políticas de Threat Prevention")

rules_metadata_threat_prevention_res = he.policy_general_api.get_all_rules_metadata(
    header_params={ "x-mgmt-run-as-job": 'on'},
    query_params={"ruleFamily": 'Threat Prevention'})


rules_metadata_threat_prevention = rules_metadata_threat_prevention_res.payload
rules_tp = []
for item in rules_metadata_threat_prevention:
    record = {
        "Orientation": item["orientation"],
        "ID": item["id"],
        "Name": item["name"],
        "Connection State": item["connectionState"],
        "Family": item["family"],
        "Is Default Rule": item["isDefaultRule"],
        "Order": item["order"]
    }
    rules_tp.append(record)

df_rules_tp = pd.DataFrame(rules_tp)
df_rules_tp
```

```python
id = df_rules_tp["ID"][3]

threat_prevention_policy_res = he.policy_threat_prevention_api.get_threat_prevention_policy(
        header_params={ "x-mgmt-run-as-job": 'on'},
        query_params={"detailsLevel": "NORMAL"},
        path_params={"ruleId": id}
    )

threat_prevention_policy = threat_prevention_policy_res.payload
print("Threat prevention for policy", id)
threat_prevention_policy
```

```python
# Requisição de dados de vulnerabilidades para gerenciamento
print("Requisitando dados de vulnerabilidades para gerenciamento")
vulnerabilities_data_res = he.posture_management_vulnerabilities_api.get_vulnerabilities_data(
    header_params={ "x-mgmt-run-as-job": 'on'}, 
    query_params={"pageSize":1000, "offset":0}
)

vulnerabilities_data = vulnerabilities_data_res.payload
print(vulnerabilities_data)


vulnerability_data_list = []
for item in vulnerabilities_data["data"]:
    vulnerability_data = {
        "Application Name": item["application"]["name"],
        "Application Version": item["application"]["version"],
        "Device Name": item["device"]["name"],
        "Device Type": item["device"]["type"],
        "Last Device Scan Status": item["device"]["lastScanStatus"],
        "Last Device Scan": item["device"]["lastScanned"],
        "Device CVE Count": item["device"]["cveCount"],
        "Device Vulnerable Products Count": item["device"]["vulnerableProductsCount"],
        "Vulnerability CVE ID": item["vulnerability"]["cveId"],
        "Vulnerability CVSS Score": item["vulnerability"]["cvssScore"],
        "Vulnerability Description": item["vulnerability"]["description"],
        "Patch File Name": ", ".join(item["patch"]["fileNames"]),
        "Patch Status": item["patch"]["status"],
        "Patch Size (MB)": item["patch"]["fileSizeMegabyteSum"]
    }
    vulnerability_data_list.append(vulnerability_data)

vulnerabilities_data_df = pd.DataFrame(vulnerability_data_list)

vulnerabilities_data_df
```
